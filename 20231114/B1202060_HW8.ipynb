```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# 函數從網址加載圖片
def load_image(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

# 定義變換
transforms_dict = {
    "Original Image": None,
    "Resized": transforms.Resize((1080, 1080)),
    "Center Cropped": transforms.CenterCrop(300),
    "Color Jittered": transforms.ColorJitter(brightness=0.8, contrast=1.5, saturation=1.3),
    "Horizontally Flipped": transforms.RandomHorizontalFlip(p=0.25),
    "Randomly Rotated": transforms.RandomRotation(80)
}

# 加載圖片
img_url = 'https://img2.40407.com/upload/202306/01/acc615Kdq62ZEus.jpg'
img = load_image(img_url)

# 準備繪圖
fig, ax = plt.subplots(1, 6, figsize=(15, 5))  # 1 row，6 column

# 繪製和顯示變換後的圖像
for i, (title, transform) in enumerate(transforms_dict.items()):
    transformed_img = img
    if transform:
        transformed_img = transform(transformed_img)
    ax[i].imshow(transformed_img)
    ax[i].set_title(title)
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# 函數從網址加載圖片
def load_image(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

# 定義變換
transforms_dict = {
    "Original Image": None,
    "Resized": transforms.Resize((1920, 1024)),
    "Center Cropped": transforms.CenterCrop(512),
    "Color Jittered": transforms.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5),
    "Horizontally Flipped": transforms.RandomHorizontalFlip(p=1),
    "Randomly Rotated": transforms.RandomRotation(60)
}

# 加載圖片
img_url = 'https://pic.pimg.tw/littlewu0502/1413524640-824133113.jpg'
img = load_image(img_url)

# 準備繪圖
fig, ax = plt.subplots(1, 6, figsize=(15, 5))  # 1 row，6 column

# 繪製和顯示變換後的圖像
for i, (title, transform) in enumerate(transforms_dict.items()):
    transformed_img = img
    if transform:
        transformed_img = transform(transformed_img)
    ax[i].imshow(transformed_img, cmap='gray')
    ax[i].set_title(title)
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```
